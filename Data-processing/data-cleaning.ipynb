```python
import pandas as pd

movies = pd.read_csv('dataset/movies.csv')
links = pd.read_csv('dataset/links.csv')

# Gop 2 bảng movies và links
movies_merged = pd.merge(movies, links, on='movieId', how='inner')

# Bỏ các movie thiếu tmbId
movies_merged = movies_merged.dropna(subset=['tmdbId'])
movies_merged['tmdbId'] = movies_merged['tmdbId'].astype(int)
final_movies = movies_merged[['movieId', 'title', 'genres', 'tmdbId']]
final_movies.to_csv('dataset/movies_ready_for_db.csv', index=False)

print(f"Có {len(final_movies)} bộ phim.")
```
